```python
import numpy as np
import pandas as pd
from gensim.models import KeyedVectors
```

# Cargar el modelo

```python
#1. Ruta al archivo 
archivo_modelo = "../SBW-vectors-300-min5.bin.gz"

print(f"Cargando modelo {archivo_modelo}...")
print("Esto puede tardar...")

modelo = KeyedVectors.load_word2vec_format(archivo_modelo, binary = True)

print("Modelo cargado existosamente")
print("-"*50)
```

# Funcion para encontrar similitudes

```python
def procesar_para_psicologo(lista_animales):
    datos_similitud = []
    datos_vectores = []

    for i, palabra in enumerate(lista_animales):
        palabra = palabra.lower().strip()
        try:
            vector = modelo[palabra]
            
            # 1. Preparar datos de similitud
            sim = modelo.similarity(lista_animales[i-1].lower(), palabra) if i > 0 else None
            datos_similitud.append({"Palabra": palabra, "Similitud_Coseno": sim})

            # 2. Preparar fila de vectores (Palabra + 300 dimensiones)
            fila_vec = {"Palabra": palabra}
            for idx, val in enumerate(vector):
                fila_vec[f"Dim_{idx+1}"] = val
            datos_vectores.append(fila_vec)

        except KeyError:
            print(f"Palabra '{palabra}' no encontrada.")

    # Guardar a archivos CSV
    df_sim = pd.DataFrame(datos_similitud)
    df_sim.to_csv("../resultados/resultados_similitud.csv", index=False)
    
    df_vec = pd.DataFrame(datos_vectores)
    df_vec.to_csv("../resultados/lista_de_vectores.csv", index=False)
    
    print("¡Archivos generados: 'resultados_similitud.csv' y 'lista_de_vectores.csv'!")
```

# Buscar similitud entre animales

```python
# Ejemplo de uso
animales = ["perro", "gato", "ratón", "jirafa", "pájaro", "león", "tigre"]
procesar_para_psicologo(animales)
```
